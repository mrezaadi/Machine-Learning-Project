# **LIBRARY**

```python
import time
import datetime
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

# **STRATEGI BRUTE FORCE DAN DIVIDE AND CONQUER**

```python
def bf_multiplication(A,B):
    hasil = 0
    for i in range(A):
        hasil = hasil + B
    return hasil
```

```python
def dc_multiplication(a,b):
  if a < 10 or b < 10:
      return a * b
  else:
      n = max(len(str(a)), len(str(b)))
      half = n//2

      aL = a // (10 ** (half))
      aR = a % (10 ** (half))
      bL = b // (10 ** (half))
      bR = b % (10 ** (half))

      x0 = dc_multiplication(aL,bL)
      x1 = dc_multiplication(aR,bR)
      x2 = dc_multiplication(aL+aR, bL+bR)-x0-x1
      return x0 * (10 ** (2*half)) + (x2 * (10 ** half)) + x1
```

# **RUNNING TIME**

## BRUTE FORCE

```python
a = [1000, 10000, 100000, 1000000, 10000000, 100000000, 1000000000]
b = [1000, 10000, 100000, 1000000, 10000000, 100000000, 1000000000]

all_rn = []
for j in range(3):
  runtime = []
  for i in range(len(a)):
    start = time.time()
    bf_multiplication(a[i], b[i])
    runtime.append(time.time() - start)
  
  plt.plot(np.arange(0,len(runtime),1),runtime,label=f'run ke-{j+1}');
  all_rn.append(np.sum(runtime))
plt.legend();
plt.title("Grafik Running Time BF")
plt.xlabel("Data index ke-")
plt.ylabel("Waktu (detik)")
```

**Grafik sebenarnya linear, tetapi karena selisih nilai yang cukup besar sehingga perubahan running time nya juga besar. Jika menggunakan range yang sama, maka akan terlihat seperti yang dibawah ini.**
