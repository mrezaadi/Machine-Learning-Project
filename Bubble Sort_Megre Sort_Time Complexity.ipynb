Bubble sort is O(n2) and Merge Sort is O(nlogn). So, on a small set it wont be that different but on a lot of data Bubble sort will be much slower.

BigO.png

```python
import matplotlib.pyplot as plt
import numpy as np
import timeit
import math
import random
from matplotlib.lines import Line2D


plt.rcParams['figure.figsize'] = [15, 7] # set size of plot
```

**MERGE SORT**

BigO(n*logn)

```python
def merge_sort(arr):
    if len(arr) > 1:
  
         # Finding the mid of the array
        mid = len(arr)//2
  
        # Dividing the array elements
        L = arr[:mid]
  
        # into 2 halves
        R = arr[mid:]
  
        # Sorting the first half
        merge_sort(L)
  
        # Sorting the second half
        merge_sort(R)
  
        i = j = k = 0
  
        # Copy data to temp arrays L[] and R[]
        while i < len(L) and j < len(R):
            if L[i] < R[j]:
                arr[k] = L[i]
                i += 1
            else:
                arr[k] = R[j]
                j += 1
            k += 1
  
        # Checking if any element was left
        while i < len(L):
            arr[k] = L[i]
            i += 1
            k += 1
  
        while j < len(R):
            arr[k] = R[j]
            j += 1
            k += 1

ns = np.linspace(0, 1000, 50, dtype=int)
ts = [timeit.timeit('merge_sort(lst)',
                    setup='lst=list(range({})); random.shuffle(lst)'.format(n),
                    globals=globals(),
                    number=1)
         for n in ns]
plt.plot(ns, ts, 'or');


degree = 4
coeffs = np.polyfit(ns, ts, degree)
p = np.poly1d(coeffs)
plt.plot(ns, [p(n) for n in ns], '-r')

plt.title('Merge Sort (n*logn)', loc='center',pad=20, fontsize=15, color='black')
plt.xlabel('Input Size (0-1000)', fontsize=10)
plt.ylabel('Time (second)', fontsize=10)
plt.grid(color='black', linestyle=':', linewidth=0.5)
```

**BUBBLE SORT**

BigO(n)

```python
def bubble_sort(our_list):
    # We go through the list as many times as there are elements
    for i in range(len(our_list)):
        # We want the last pair of adjacent elements to be (n-2, n-1)
        for j in range(len(our_list) - 1):
            if our_list[j] > our_list[j+1]:
                # Swap
                our_list[j], our_list[j+1] = our_list[j+1], our_list[j]
# 15 values
ns = np.linspace(0, 1000, 50, dtype=int)
ts = [timeit.timeit('bubble_sort(lst)',
                    setup='lst=list(range({})); random.shuffle(lst)'.format(n),
                    globals=globals(),
                    number=1)
         for n in ns]
plt.plot(ns, ts, 'or', color="blue");

degree = 4
coeffs = np.polyfit(ns, ts, degree)
p = np.poly1d(coeffs)
plt.plot(ns, [p(n) for n in ns], '-r', color="blue")

plt.title('Bubble Sort (n^2)', loc='center',pad=20, fontsize=15, color='black')
plt.xlabel('Input Size (0-1000)', fontsize=10)
plt.ylabel('Time (second)', fontsize=10)
plt.grid(color='black', linestyle=':', linewidth=0.5)
```
